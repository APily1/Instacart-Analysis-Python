### Creating new variable

```python
# using 'def' with conditional function 'if, elif, else'.
def price_label(row):
    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High-range'
    else: 
        return 'Not enough data'
```

```python
df['price_range'] = df.apply(price_label, axis = 1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
df['prices'].max()
```

### If-Statements using the loc() function

```python
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Min-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df['price_range_loc'].value_counts(dropna = False)
```

```python
df_ords_prods.loc[(df_ords_prods['prices'] <= 15) & (df_ords_prods['prices'] > 5), 'price_range_loc'] = 'Min-range product'

df_ords_prods.loc[df_ords_prods['prices'] > 15, 'price_range_loc'] = 'High-range product'

df_ords_prods.loc[df_ords_prods['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

### If-Statments with the for-loops

```python
import time
start_time = time.time()

result = []

for value in df_ords_prods["order_dow"]:
  if value == 'Saturday':
    result.append("Busiest day")
  elif value == 'Wednesday':
    result.append("Least busy")
  else:
    result.append("Regularly busy")
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
result
```

```python
df_ords_prods['busiest_day'] = result
```

```python
# Checking frequency
df_ords_prods['busiest_day'].value_counts(dropna = False)
```

### Creating a "busiest_days" column

```python
start_time = time.time()

result2 = []

for value in df_ords_prods["order_dow"]:
  if value == 'Saturday' or value == 'Sunday':
    result2.append("Busiest days")
  elif value == 'Wednesday' or value =='Tuesday':
    result2.append("Slowest days")
  else:
    result2.append("Regularly busy")
print("--- %s seconds ---" % (time.time() - start_time))
```

***Adding result to the new column "busiest_days"

```python
df_ords_prods['busiest_days'] = result2
```

```python
df_ords_prods['busiest_days'].value_counts(dropna = False)
```

## Identifying the busiest hours of the day

```python
# Finding busiest hour
df_ords_prods['order_hour_of_day'].value_counts(dropna = False)
```

*** Labeling "most orders, average orders, and fewest orders" then creating a new column called "busiest_period_of_day"

```python
result_per_hours = []

for value in df_ords_prods["order_hour_of_day"]:
  if value >= 9 and value <= 16:
    result_per_hours.append("Most orders")
  elif value == 23 or (value <= 6 and value >= 0):
    result_per_hours.append("Fewest orders")
  else:
    result_per_hours.append("Average orders")
```

```python
df_ords_prods['busiest_period_of_day'] = result_per_hours
```
